```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def distance(data,cluster):
    dic_cluster={}
    k=1
    for i in cluster:
        l=[]
        for j in data:
            dis=math.sqrt((i[0]-j[0])**2+(i[1]-j[1])**2)
            l.append(dis)
        dic_cluster[f"c{k}"]=l
        k+=1
    return dic_cluster
```

```python
def find_centroid(cluster):
    cen=[]
    x=[i[0] for i in cluster]
    y=[i[1] for i in cluster]
    cen.append(round(sum(x)/len(x)))
    cen.append(round(sum(y)/len(y)))
    return cen
```

```python
data=[[1,2],[2,6],[12,10],[3,10],[3,8],[14,9],[4,6],[15,10],
[6,9],[6,11],[6,5],[7,5],[15,4],[7,2],[15,18],[4,7]]
cluster=[[1,2],[2,6],[12,10]]
cluster2=[[1,2],[2,6]]

print("K Means Clustering K=3 Simulation ......")
print()
while(1):
    dis_list=distance(data,cluster)
    print(f"The Distance :",dis_list)
    centroid_list,c1,c2,c3=assign(dis_list)
    print(f"The New Clusters : {centroid_list}")
    convergence = all(cluster[i] == centroid_list[i] for i in range(len(cluster)))
    if convergence:
        print("Convergence Achieved\n")
        break
    cluster=centroid_list
print("Plot The Clusters.....")
plot(data,centroid_list,c1,c2,c3)

print("K Means Clustering K=2 Simulation ......")
print()

while(1):
    dis_list=distance(data,cluster2)
    print(f"The Distance :",dis_list)
    centroid_list,c1,c2=assign2(dis_list)
    print(f"The New Clusters : {centroid_list}")
    convergence = all(cluster2[i] == centroid_list[i] for i in range(len(cluster2)))
    if convergence:
        print("Convergence Achieved\n")
        break
    cluster2=centroid_list
print("Plot The Clusters.....")
plot2(data,centroid_list,c1,c2)
```

```python
def plot(data, centroid, c1, c2, c3):
    # Convert clusters to numpy arrays
    c1 = np.array(c1)
    c2 = np.array(c2)
    c3 = np.array(c3)
    centroid = np.array(centroid)
    
    plt.figure(figsize=(10, 6))
    
    # Plot the clusters
    sns.scatterplot(x=c1[:, 0], y=c1[:, 1], color='red', label='Cluster 1')
    sns.scatterplot(x=c2[:, 0], y=c2[:, 1], color='blue', label='Cluster 2')
    sns.scatterplot(x=c3[:, 0], y=c3[:, 1], color='green', label='Cluster 3')
    
    # Plot centroids
    sns.scatterplot(x=centroid[:, 0], y=centroid[:, 1], color='black', marker='X', s=100, label='Centroids')
    
    plt.xlabel('X-coordinate')
    plt.ylabel('Y-coordinate')
    plt.title('K-Means Clustering Results')
    plt.legend()
    plt.show()
```
